Building and training neural networks is a core aspect of deep learning, where neural networks are designed to recognize patterns and perform tasks such as classification, regression, and clustering. 

Below is an overview of the steps involved in constructing and training a neural network:

1. Building Neural Networks<br>
a. Architecture Design<br>
- Input Layer: 
    - This layer represents the input features of the dataset. 
    - The number of neurons in the input layer is equal to the number of input features.
- Hidden Layers: 
    - These layers sit between the input and output layers and consist of neurons that learn complex patterns. 
    - The number of hidden layers and neurons in each layer is determined through experimentation.
- Output Layer: 
    - This layer provides the final prediction. 
    - For classification, it can have softmax activation for multi-class problems, or sigmoid for binary classification.<br>

b. Choosing Activation Functions
- Sigmoid: Maps values between 0 and 1, often used for binary classification tasks.
- ReLU (Rectified Linear Unit): Most commonly used for hidden layers, ReLU introduces non-linearity and helps with faster convergence.
- Tanh: Like Sigmoid but outputs values between -1 and 1, used less frequently but sometimes effective for certain datasets.<br>


c. Weights and Biases Initialization
- Weights are initialized randomly or using advanced methods like He initialization or Xavier initialization to improve convergence.
- Biases are usually initialized to small positive values or zero.<br>

d. Loss Function
- Mean Squared Error (MSE) for regression tasks.
- Cross-Entropy Loss for classification tasks, particularly useful when dealing with probabilities.<br>

e. Optimizer
- Gradient Descent and its variants are used to minimize the loss function. The most common optimizers include:
    - Stochastic Gradient Descent (SGD): Updates parameters using a subset of the training data.
    - Adam: An adaptive learning rate optimizer that is widely used because of its faster convergence and robustness.<br>

<b>2. Training Neural Networks</b><br>
a. Forward Propagation
- During forward propagation, input data is passed through the network. Each layer applies the activation function to the weighted sum of inputs to produce output, which is then passed to the next layer.<br>

b. Backpropagation
- Backpropagation adjusts weights and biases to minimize the loss function. It computes the gradient of the loss function concerning each weight by applying the chain rule and propagating errors backward from the output layer to the input layer.<br>

c. Batching Data
- Batch Gradient Descent: Entire training data is used for each update.
- Mini-Batch Gradient Descent: Small batches of the data are used, making the process faster and more memory efficient.<br>

d. Regularization Techniques
- Dropout: Randomly sets a fraction of neurons to zero during training to prevent overfitting.
- L1 and L2 Regularization: Adds a penalty term to the loss function to constrain model complexity.<br>

e. Training Process
- Initialize parameters (weights and biases).
- For each epoch (complete pass over the training data):
- Perform forward propagation to compute predictions.
- Compute the loss using the loss function.
- Perform backpropagation to compute the gradients.
- Update weights and biases using the optimizer.
- Continue until the model converges (loss no longer significantly decreases).<br>

<b>3. Evaluation</b><br>
- After training, the model is evaluated on unseen data (test set) to measure performance, typically using metrics like accuracy for classification tasks or mean absolute error for regression tasks.

- What Are Activation Functions?
    - Activation functions are an integral building block of neural networks that enable them to learn complex patterns in data. 
    - They transform the input signal of a node in a neural network into an output signal that is then passed on to the next layer. 
    - Without activation functions, neural networks would be restricted to modeling only linear relationships between inputs and outputs. 
    - Activation functions introduce non-linearities, allowing neural networks to learn highly complex mappings between inputs and outputs. 

<br>
- Choosing the right activation function is crucial for training neural networks that generalize well and provide accurate predictions.
<br>
- Understanding activation functions in depth will build your intuition and improve your application of neural networks.
<br>


<b>Types of Activation Functions</b>

- The linear activation function is the simplest activation function, defined as:<br>

    - f(x) = x
<br>
- It simply returns the input x as the output. Graphically, it looks like a straight line with a slope of 1.<br>

- The main use case of the linear activation function is in the output layer of a neural network used for regression. <br>
- For regression problems where we want to predict a numerical value, using a linear activation function in the output layer ensures the neural network outputs a numerical value.  <br>
- The linear activation function does not squash or transform the output, so the actual predicted value is returned.<br>
<br>
- NOTE: 
    - However, the linear activation function is rarely used in hidden layers of neural networks. This is because it does not provide any non-linearity. The whole point of hidden layers is to learn non-linear combinations of the input features. Using a linear activation throughout would restrict the model to just learning linear transformations of the input.






Sigmoid activation

- The sigmoid activation function, often represented as σ(x), is a smooth, continuously differentiable function that is historically important in the development of neural networks. 
- The sigmoid activation function has the mathematical form:

    - f(x) = 1 / (1 + e^-x)
<br>

- It takes a real-valued input and squashes it to a value between 0 and 1. <br>
- The sigmoid function has an "S"-shaped curve that asymptotes to 0 for large negative numbers and 1 for large positive numbers. <br>
- The outputs can be easily interpreted as probabilities, which makes it natural for binary classification problems.
<br>
- Sigmoid units were popular in early neural networks since the gradient is strongest when the unit's output is near 0.5, allowing efficient backpropagation training. <br>
- However, sigmoid units suffer from the "vanishing gradient" problem that hampers learning in deep neural networks.
<br>
- As the input values become significantly positive or negative, the function saturates at 0 or 1, with an extremely flat slope. <br>
- In these regions, the gradient is very close to zero.<br> 
- This results in very small changes in the weights during backpropagation, particularly for neurons in the earlier layers of deep networks, which makes learning painfully slow or even halts it. <br>
- This is referred to as the vanishing gradient problem in neural networks.<br>
<br>

- The main use case of the sigmoid function is as the activation for the output layer of binary classification models. <br>
- It squashes the output to a probability value between 0 and 1, which can be interpreted as the probability of the input belonging to a particular class.<br>




Tanh (hyperbolic tangent) activation

- The tanh (hyperbolic tangent) activation function is defined as:

    - f(x) = (e^x - e^-x) / (e^x + e^-x)
    
<br>

- The tanh function outputs values in the range of -1 to +1. <br>
- This means that it can deal with negative values more effectively than the sigmoid function, which has a range of 0 to 1. <br>

- Unlike the sigmoid function, tanh is zero-centered, which means that its output is symmetric around the origin of the coordinate system. <br>
- This is often considered an advantage because it can help the learning algorithm converge faster.<br>

- Because the output of tanh ranges between -1 and +1, it has stronger gradients than the sigmoid function. <br>
- Stronger gradients often result in faster learning and convergence during training because they tend to be more resilient against the problem of vanishing gradients when compared to the gradients of the sigmoid function.<br>

- Despite these advantages, the tanh function still suffers from the vanishing gradient problem. <br>
- During backpropagation, the gradients of the tanh function can become very small (close to zero). <br>
- This issue is particularly problematic for deep networks with many layers; the gradients of the loss function may become too small to make significant changes in the weights during training as they propagate back to the initial layers. <br>
- This can drastically slow down the training process and can lead to poor convergence properties.<br>

- The tanh function is frequently used in the hidden layers of a neural network. <br>
- Because of its zero-centered nature, when the data is also normalized to have mean zero, it can result in more efficient training.<br>

- NOTE: 
    - If one has to choose between the sigmoid and tanh and has no specific reason to prefer one over the other, tanh is often the better choice because of the reasons mentioned above. However, the decision can also be influenced by the specific use case and the behavior of the network during initial training experiments.

- Let’s dive into the detailed explanation of building and training a neural network for customer churn prediction using synthetic data.

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Create synthetic data for customer churn prediction
np.random.seed(42)  # For reproducibility

# Number of samples
n_samples = 1000

# Generating random features (age, monthly charges, tenure, contract type, support tickets)
age = np.random.randint(18, 70, n_samples)
# monthly_charges = np.random.uniform(20, 100, n_samples)
# monthly_charges = np.round(np.random.uniform(20,100, n_samples),2)
 
tenure = np.random.randint(1, 60, n_samples)  # in months
contract_type = np.random.choice([0, 1], size=n_samples)  # 0: Month-to-month, 1: Annual
support_tickets = np.random.randint(0, 10, n_samples)

# Randomly generating target labels (Churn: 0 = No, 1 = Yes) based on feature patterns
# We assume higher monthly charges, lower tenure, and more support tickets lead to more churn.
churn_prob = (monthly_charges - np.mean(monthly_charges)) + (10 - tenure/6) + support_tickets
churn = (churn_prob > np.median(churn_prob)).astype(int)  # High churn_prob leads to churn = 1

# Create a DataFrame
df = pd.DataFrame({
    'age': age,
    'monthly_charges': monthly_charges,
    'tenure': tenure,
    'contract_type': contract_type,
    'support_tickets': support_tickets,
    'churn': churn
})

df.head()  # Show the first few rows of synthetic data
```

```python
# Splitting features and target
X = df.drop('churn', axis=1).values
y = df['churn'].values

# Split the dataset into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalize the features using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Build the neural network model
model = Sequential()

# Input layer and first hidden layer (32 neurons)
model.add(Dense(32, input_dim=X_train.shape[1], activation='relu'))

# Second hidden layer (16 neurons)
model.add(Dense(16, activation='relu'))

# Dropout layer (prevents overfitting by randomly setting some neurons to 0 during training)
model.add(Dropout(0.5))

# Output layer (sigmoid activation for binary classification)
model.add(Dense(1, activation='sigmoid'))

# Compile the model (binary_crossentropy for binary classification, adam for optimization)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Summary of the model architecture
model.summary()
```

```python
# Train the model for 50 epochs and use a batch size of 32
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Plotting the accuracy over epochs
plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='upper left')
plt.show()
```
