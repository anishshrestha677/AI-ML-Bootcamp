### What is LSTM?
- [Full Form] LSTM stands for Long Short-Term Memory
- [Algo Type] Is a specialized type of RNN 
- Is capable of learning long-term dependencies in sequential data
- making it highly suitable for time series prediction.
- LSTMs use "gates" (input, forget, and output gates) to selectively remember or forget data points over time.
- Gates allow LSTMs to store information about previous time steps and make better predictions by considering both recent and long-term trends.
